```python
import cv2 import cv2

# 圖片檔案名稱
image_path = 'fcu.jfif'
output_path_2 = 'grayscale_direct_load.jfif'

# 使用旗標 0 讀取圖片，表示直接以灰階模式載入
grayscale_image_direct = cv2.imread(image_path, 0) # 0 或 cv2.IMREAD_GRAYSCALE

if grayscale_image_direct is None:
    print(f"錯誤：無法載入圖片。請檢查路徑：{image_path}")
else:
    # 儲存灰階圖片
    cv2.imwrite(output_path_2, grayscale_image_direct)
    print(f"直接讀取的灰階圖片已儲存為：{output_path_2}")

    # 顯示圖片
    cv2.imshow('Direct Grayscale Load', grayscale_image_direct)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import numpy as np
from PIL import Image

import matplotlib.pyplot as plt

def redchannel(imagename):
  # 1. Pillow 讀取圖片
  img = Image.open('fcu.jpg')

  # 2. 轉為 NumPy 陣列
  img_array = np.array(img)

  # 3. NumPy 分離通道
  red_channel_img = img_array.copy()
  red_channel_img[:, :, 1] = 0  # G=0
  red_channel_img[:, :, 2] = 0  # B=0

  # 4. Matplotlib 顯示 (此處由 JS 模擬)
  plt.imshow(red_channel_img)

def greenchannel(imagename):
  # 1. Pillow 讀取圖片
  img = Image.open('fcu.jpg')

  # 2. 轉為 NumPy 陣列
  img_array = np.array(img)

  # 3. NumPy 分離通道
  green_channel_img = img_array.copy()
  green_channel_img[:, :, 0] = 0  # R=0
  green_channel_img[:, :, 2] = 0  # B=0

  # 4. Matplotlib 顯示 (此處由 JS 模擬)
  plt.imshow(green_channel_img)

def bluechannel(imagename):
  # 1. Pillow 讀取圖片
  img = Image.open('fcu.jpg')

  # 2. 轉為 NumPy 陣列
  img_array = np.array(img)

  # 3. NumPy 分離通道
  blue_channel_img = img_array.copy()
  blue_channel_img[:, :, 0] = 0  # R=0
  blue_channel_img[:, :, 1] = 0  # G=0

  # 4. Matplotlib 顯示 (此處由 JS 模擬)
  plt.imshow(blue_channel_img)
```

```python
def maxiumum(value1,value2,value3):
  """return the maximum of the three values."""
  max_value = value1
  if value2 > max_value:
    max_value = value2
  if value3 > max_value:
    max_value = value3
  return max_value
```

```python
maxvalue=maxiumum(100,1,99)
```

```python
def miniumum(value1,value2,value3,value4):
  """return the minimum of the four values."""
  min_value = value1
  if value2 > min_value:
    min_value = value2
  if value3 > min_value:
    min_value = value3
  if value4 > min_value:
    min_value = value4
  return min_value
```

```python
miniumum(1,2,3,4)
```
